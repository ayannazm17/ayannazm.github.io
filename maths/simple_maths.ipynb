Create Database

```python
import sqlite3
import pandas as pd

# Create or connect to a database file
conn = sqlite3.connect("shop.db")
cur = conn.cursor()

print("Database connected successfully.")
```

Create simple Table

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS sales (
    sale_id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_name TEXT,
    category TEXT,
    price REAL,
    quantity INTEGER
);
""")

conn.commit()
print("Table created.")
```

Insert sample data

```python
cur.executemany("""
INSERT INTO sales (product_name, category, price, quantity)
VALUES (?, ?, ?, ?);
""", [
    ('Laptop', 'Electronics', 800, 2),
    ('Phone', 'Electronics', 500, 3),
    ('Headphones', 'Electronics', 100, 5),
    ('Desk Chair', 'Furniture', 150, 4),
    ('Desk', 'Furniture', 300, 2),
    ('Notebook', 'Stationery', 5, 20),
    ('Pen Pack', 'Stationery', 10, 10)
])

conn.commit()
print("Data inserted.")
```

View all data

```python
df = pd.read_sql_query("SELECT * FROM sales ", conn)
df
```

Calculate the AVG Price

```python
pd.read_sql_query("""
SELECT AVG(price) FROM sales
""", conn)
```

Calculate Total Quantity Sold

```python
pd.read_sql_query("""
SELECT SUM(quantity) FROM sales
""", conn)
```

Calculate Highest and Lowest Price

```python
pd.read_sql_query("""
SELECT MAX(price), MIN(price) FROM sales
""", conn)
```

Calculate revenue per product

```python
pd.read_sql_query("""
SELECT product_name, price * quantity AS revenue FROM sales
""", conn)
```

Calculate total revenue

```python
pd.read_sql_query("""
SELECT SUM(price * quantity) FROM sales
""", conn)
```

Calculate revenue per category

```python
pd.read_sql_query("""
SELECT category, SUM(price * quantity) AS revenue FROM sales
GROUP BY category
""", conn)
```

Calculate average price per category

```python
pd.read_sql_query("""
SELECT category, AVG(price) AS average_price FROM sales
GROUP BY category
""", conn)
```

Calculate count products per category

```python
pd.read_sql_query("""
SELECT category, COUNT(*) AS product_count FROM sales
GROUP BY category
""", conn)
```

Which category has a revenue greater than 2000

```python

pd.read_sql_query("""
SELECT category, SUM(price * quantity) AS revenue FROM sales
GROUP BY category
HAVING revenue > 2000
""", conn)
```
